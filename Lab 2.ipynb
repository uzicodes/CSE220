```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))

```

Task 1: Zigzag Walk

```python
def walk_zigzag(floor):
  row, column = floor.shape
  for index1 in range(row):
    if index1%2 == 0:
      for index2 in range(column):
        if index2%2 == 0:
          print(floor[index1][index2], end = " ")
      print()
    else:
      for index2 in range(column-1, -1, -1):
        if index2%2 != 0:
          print(floor[index1][index2], end = " ")
      print()

floor = np.array([[ '3W' , '8B' , '4W' , '6B' , '1W' , '5B'],['3B' , '2W' , '1B' , '6W' , '3B' , '8W'],
                  ['9W' , '0B' , '7W' , '5B' , '3W' , '8B'],['2B' , '1W' , '3B' , '6W' , '0B' , '4W'],
                   ['1W' , '4B' , '2W' , '8B' , '6W' , '6B']])
print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3W 4W 1W
# 8W 6W 2W
# 9W 7W 3W
# 4W 6W 1W
# 1W 2W 6W
```

Task 2: Landscape Screen

```python
def landscape(m):
  row, column = m.shape
  final_array = np.zeros((column, row), dtype = int)
  for index1 in range(row):
    for index2 in range(column):
      final_array[index2][index1] = m[index1][index2]
  return final_array

m = np.random.randint(15,size=(4,3)) #This creates an np array of 4 x 3 where the values are random numbers from [0,15) range
print_matrix(m)
print()
print_matrix(landscape(m))
```

Task 3: Seating Arrangement

```python
def arrangement(m, diff):
  row, column = m.shape
  for index1 in range(row):
    for index2 in range(column):
      if index1 != index2:
        if abs(m[index1][index2] - m[index2][index1]) != diff:
          return False
  return True

m = np.array([[7,13,9,14],[12,8,15,11],[10,17,3,2],[15,10,1,4]])
print_matrix(m)
print()
returned_value = arrangement(m,1)
print(returned_value)#This should print False
unittest.output_test(returned_value, False)
```

Task 4: Chess Piece

```python
def show_knight_move(knight):
  knight_move = np.zeros((8,8), dtype = int)
  row, column = knight

  knight_move[row][column] = 33
  possible_positions = np.array([[2, 1], [2, -1], [-2, 1], [-2, -1], [1, 2], [1, -2], [-1, 2], [-1, -2]])

  for each in possible_positions:
    vertical = row + each[0]
    horizontal = column + each[1]
    if vertical >= 0 and vertical <= 7 and horizontal >= 0 and horizontal <= 7:
      knight_move[vertical][horizontal] = 3

  return knight_move

knight = (3, 4)
chess_board = show_knight_move(knight)
print_matrix(chess_board)
```

Bonus Task

```python
def topple(knight,rook):
  knight_move = np.zeros((8,8), dtype = int)
  row, column = knight

  knight_move[row][column] = 33
  possible_positions = np.array([[2, 1], [2, -1], [-2, 1], [-2, -1], [1, 2], [1, -2], [-1, 2], [-1, -2]])

  for each in possible_positions:
    vertical = row + each[0]
    horizontal = column + each[1]
    if vertical >= 0 and vertical <= 7 and horizontal >= 0 and horizontal <= 7:
      knight_move[vertical][horizontal] = 3

  knight_move[rook[0]][rook[1]] = 5
  rook_vertical = np.array([[0, rook[1]]]*8)
  rook_horizontal = np.array([[rook[0], 0]]*8)

  index1 = 0
  for each in rook_vertical:
    each[0] = index1
    index1 += 1

  index2 = 0
  for each in rook_horizontal:
    each[1] = index2
    index2 += 1

  for each in rook_vertical:
    if each[0] == row and each [1] == column:
      return (knight_move, "Rook can kill")

  for each in rook_horizontal:
    if each[0] == row and each [1] == column:
      return (knight_move, "Rook can kill")

  return (knight_move, "Cannot Kill")

knight = (5,6)
rook = (5,1)
chess_board, ans = topple(knight,rook)
print_matrix(chess_board)
print(ans) #This should print Rook can kill
```
